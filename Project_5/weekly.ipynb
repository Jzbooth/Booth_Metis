```python
def to_matrix(string):
    string = string.replace('[','')
    string = string.replace(']','')
    string = string.replace('(','')
    string = string.replace(')','')
    string = string.replace(',','')
    string = string.replace("'",'')
    string = string.replace('array','')
    string = string.replace('\n',' ')
    string = string.split('WEEK')
    
    matrix = list()
    for week in string[1:]:
        matrix.append(week.split())
    return matrix
```

```python
def to_df(matrix):
# create a list of lists where each item is the player name, points scored, week, and 1 if prediction 
# 0 if real
    new_list = []
    i = 0
    for week in matrix:
        i += 1
        k = 2
        pred_real = 1
        while k < len(week):
            if week[k] == 'True':
                k += 1
                if k >=len(week):
                    continue
                pred_real = 0  
            x = [week[k] + ' ' + week[k+1], week[k+2], i, pred_real]
            new_list.append(x)
            k += 3
    return new_list
```
