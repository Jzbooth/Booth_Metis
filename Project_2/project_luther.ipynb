```python
from __future__ import print_function, division
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
```

```python
url = 'http://www.spotrac.com/mlb/rankings/2017/' #webiste containing the data

response = requests.get(url)
```

```python
response.status_code
```

```python
print(response.text)
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page,"lxml")
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
tables=soup.find_all("table")
rows=[row for row in tables[0].find_all('tr')]

# Just want to look at 1st 20 rows for now
rows=rows[1:10000000]

salaries={}
for row in rows:
    items=row.find_all('td')
    player=items[1].find('a')['href']
    salaries[player]=[i.text for i in items[1:]]
    

list(salaries.items())[0][1][1]
salaries['http://www.spotrac.com/redirect/player/5273/'][1]
```

```python
# why is it only taking the first 100 rows???
count = 0
for player in salaries:
    count = count + 1
print(count)
```

```python
def money_to_int(moneystring):
    moneystring = moneystring.replace('$', '').replace(',', '').replace(' ','')
    return int(moneystring)
```

```python
money_to_int(list(salaries.items())[0][1][2])
```

```python
def getname(namestring):
    namestring = namestring.split()
    return(namestring[0] + ' ' + namestring[1])
```

```python
getname(list(salaries.items())[0][1][1])
```

```python
salarydict = defaultdict(list)
for player in salaries:
    salarydict[getname(salaries[player][1])] = money_to_int(salaries[player][2])
    
    
    
```
